```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

## Data Collection and Preprocessing

```python
## loading the data
parkinson_data = pd.read_csv('Parkinsson-Disease-Model/Parkinsson disease.csv')
```

```python
## printing the first 10 rows
parkinson_data.head(10)
```

```python
## shape of the data
parkinson_data.shape
```

```python
# 5 pointer summary
parkinson_data.describe()
```

```python
## basic info about data
parkinson_data.info()
```

```python
parkinson_data['status'].value_counts()
```

1 --> Parkinsson's Positive, 0 --> Healthy

```python
#checking for missing values in each column
parkinson_data.isnull().sum()
```

```python
## separating the dependent and independent features
x = parkinson_data.drop(columns=['status','name'])
y = parkinson_data['status']
```

## Standardize the data

```python
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
x_std
```

## Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(x_std,y,test_size = 0.2 , stratify=y , random_state=4)
```

```python
x_train.shape , x_test.shape
```

## Training the model

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

## Model Evaluation

### Accuracy Score

```python
## accuracy score on the training data
pred_train = model.predict(x_train)
accuracy_score(pred_train,y_train)
```

```python
## accuracy score on the test data
pred_test = model.predict(x_test)
accuracy_score(pred_test,y_test)
```

## Build a Predictive System

```python
input_data = (223.365,238.987,98.664,0.00264,0.00001,0.00154,0.00151,0.00461,0.01906,0.165,0.01013,0.01296,0.0134,0.03039,0.00301,26.138,0.447979,0.686264,-7.293801,0.086372,2.32156,0.098555)

## changing the input data to numpy array
input_data_array = np.asarray(input_data)

## reshaping the numpy array
input_data_reshape = input_data_array.reshape(1,-1)

## standardizing the data
std_data = scaler.transform(input_data_reshape,)

## making the prediction
prediction = model.predict(std_data)

print(prediction)
if prediction == 0:
    print("The Person does not have Parkinsson's Disease")
else:
    print("The Person does have Parkinsson's Disease")
```
